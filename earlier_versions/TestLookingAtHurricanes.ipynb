```python
monthStartDays = np.cumsum([0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30])
print(monthStartDays)
monthStartDaysLeap = np.cumsum([0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30])
print(monthStartDaysLeap)
```

```python
#Function to convert date and time to a decimal number o days since start of year, starting at 1 for January 1st
def yearday(date, time):
    datestring = str(date)
    #print(datestring)
    year = int(datestring[0:4])
    isLeap = np.mod(year,4) == 0
    if year == 1900:
        isLeap = false
    month = int(datestring[4:6])
    day = int(datestring[6:8])
    if not isLeap:
        yd = day + monthStartDays[month-1] #monthStartDays starts at 0 so this will start at 1
    else:
        yd = day + monthStartDaysLeap[month-1]
    ydPlusTime = yd + time/2400.
    return ydPlusTime
    
    
```

```python
#OK, now add a "yearday" column to the dataframe object.
#May not be easy to vectorize this--function written to operate on individual integers rather than a list of them
def addYeardayColumn(df):
    yeardays = np.zeros(len(df))
    for i in range(len(df)):
        yeardays[i] = yearday(df['date'][i], df['time'][i])
    df['yearday'] = yeardays
        
```

```python
#First need colors for the categories. 
#Need to calculate what category it is...
def category(wind):
    if wind < 34:
        return 'd' #Or should be a number like -1 for TD and 0 for TS?
    elif wind < 64:
        return 's'
    elif wind < 83:
        return 1
    elif wind < 96:
        return 2
    elif wind < 113:
        return 3
    elif wind < 137:
        return 4
    else:
        return 5
```

```python
#Function for category colors
def categoryColors(wind):
    if wind < 34:
        return 'turquoise' #TD
    elif wind < 64:
        return 'green' #TS
    elif wind < 83:
        return 'yellow' #Cat 1
    elif wind < 96:
        return 'orange' #Cat 2
    elif wind < 113:
        return 'red' #Cat 3
    elif wind < 137:
        return 'magenta' #Cat 4
    else: 
        return 'black' #Cat 5
```

```python
#OK, barh should work--can control width and keeps edges exactly where they should be. 
#
#Write a function to plot the bar for the individual storm.
#Then a wrapper to loop through it for all the storms in a season, add month markings, storm names, etc.
```

```python
def hurrbar(df, position, thickness):
    for i in range(len(df)):
        if df['type'][i] in ['TD', 'TS', 'HU', 'SS']: #Only plot tropical or subtropical cyclones, not remnant lows or ETs
            if i == len(df)-1: #have reached end of storm--plot fixed length of 6 hours
                plt.barh(position, 0.25, height=thickness, left=df['yearday'][i], 
                         color=categoryColors(df['wind'][i]))
            else: #plot this color until the next advisory
                plt.barh(position, df['yearday'][i+1] - df['yearday'][i], height=thickness, left=df['yearday'][i],
                         color=categoryColors(df['wind'][i]))
```

```python
plt.figure()
hurrbar(df_irma, 9, 1)
plt.axis([200, 300, 0, 20])
plt.show()
```

```python
#OK, now we're getting somewhere!
#Next step: loop through all the storms in the season.
#... also better colorbar?
```

```python
def plotSeason(year):
    storms = os.listdir(str(year))
    i = len(storms)
    plt.axis([0, 366, 0, len(storms)+1])
    for storm in storms:
        df = pd.read_csv(str(year)+'/'+storm, skiprows=1, skipinitialspace=True)
        addYeardayColumn(df)
        hurrbar(df, i, 0.9)
        i -= 1
    plt.title(year)
```
